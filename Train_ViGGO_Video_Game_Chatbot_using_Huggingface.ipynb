```python
from transformers import pipeline

qa_model = pipeline("question-answering", model="consciousAI/question-answering-roberta-base-s-v2")
question = "Are you talking about the Diablo II that has multiplayer??"
context = "confirm(name[Diablo II], has_multiplayer[yes])"
qa_model(question=question, context=context)
```
